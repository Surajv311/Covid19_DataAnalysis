### task 3.2: find maxmimum infection rate for China

```python
corona_dataset_aggregated.loc['China'].diff().max() # this would return a number 
```

```python
corona_dataset_aggregated.loc['India'].diff().max()
```

```python
corona_dataset_aggregated.loc['Australia'].diff().max()
```

### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries = list(corona_dataset_aggregated.index) # this will return the index of the list 
max_infection_rates=[]
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
max_infection_rates #now we have calculated the max inf rates in each country using a for loop
# its better to put these values as a col in the dataframe 
corona_dataset_aggregated["max_infection_rate"] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

### Task 3.4: create a new dataframe with only needed column 

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
```

### Task 5.1: Plotting GDP vs maximum Infection rate

```python
x = data["GDP per capita"] # x axis 
y = data["max_infection_rate"] # yaxis 
#sns.scatterplot(x,y) ### but this would not show proper scaling , for proper scaling we apply log() to the values
sns.scatterplot(x,np.log(y))
```
